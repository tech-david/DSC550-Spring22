```python
!pip install neurolab
import numpy as np
import matplotlib.pyplot as plt
import neurolab as nl

# hard limit instantiation
hardlim = nl.trans.HardLim()
# weights for graph a
w_a = np.array([-2, 1])
# bias for graph 1
b_a = 0
# a = 1
p_a_1 = np.array([-2,2])
# a = 0
p_a_0 = np.array([2,2])
# new point
p_a_rand = np.array([-1,1])

# calculate n for every point
n_a_1 = np.sum(np.multiply(p_a_1, w_a))+b_a
n_a_0 = np.sum(np.multiply(p_a_0, w_a))+b_a
n_a_rand = np.sum(np.multiply(p_a_rand, w_a)) + b_a

# decisioning
a_1 = hardlim(n_a_1)
a_0 = hardlim(n_a_0)
a_rand = hardlim(n_a_rand)

#results
print("input:%s --> target:%s" % (p_a_1, a_1))
print("input:%s --> target:%s" % (p_a_0, a_0))
print("input:%s --> target:%s" % (p_a_rand, a_rand))
```
