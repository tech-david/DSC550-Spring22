```python
# Only neeeds to be run once to install neurolab
#!pip install neurolab

import neurolab as nl
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Creating saturated linear variable
satlin = nl.trans.SatLin();
p = np.arange(-3, 4, 1);
plt.plot(p, satlin(p));
```

```python
# Creating linear variable
purelin = nl.trans.PureLin();
p = np.arange(-3, 4, 1);
plt.plot(p, purelin(p));
```

```python
# Creating activation function and passing it to satlin
n1_1 = 2 * p + 2;
a1_1 = satlin(n1_1);
plt.plot(p, a1_1);
```

```python
# Creating activation function and passing it to satlin
n1_2 = p - 1;
a1_2 = satlin(n1_2);
plt.plot(p, a1_2);
```

```python
# Creating activation functions and passing it to satlin then purelin activation
n1_1 = 2 * p + 2;
a1_1 = satlin(n1_1);

n1_2 = p - 1;
a1_2 = satlin(n1_2);

n2_1 = a1_1 - a1_2;
a2_1 = purelin(n2_1);
plt.plot(p, a2_1);
```
